### 1. Check whether there is any relationship between the amount  spent for advertising through five different companies and profit  generated.

#### For this, we need to identify the correlation between the profit and Advertisements

```python
corrmatrix = data.corr()
plt.subplots(figsize = (10,4))
sns.heatmap(corrmatrix, vmin = -2, vmax = 0.99, annot=True, linewidth=0.2)
```

### Insights:
#### The above Correlation Matrix shows that all the variables shows a strong positive correlation. The Advertising company 2 has the highest correlation value as 1 and the others with a correlation value 0.99

### 2. Build a regression model to predict the profit from the given data and comment your findings on it.

```python
data1=data.drop(['Date'], axis=1)
y=data1['Profit(in Thousands)']
X=data1.drop(['Profit(in Thousands)'], axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, test_size = 0.25)
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(X_train,y_train)
predictions=model.predict(X_test)
print(y_test)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE is', mean_squared_error(y_test,predictions))
```

### Insights:
##### The Test size was set as 0.25, which means that 25% of the total elements are taken for testing and 75% of the total elements are taken for Training the model. The 5.40848862053903E-025" , which is a small value, which means that the Model developed is perfect.

### 3. Predict the profit generated by the company, if they wish to spend the following amount for advertisement.
##### Add1(in Thousands) 290
##### Add2(in Thousands) 312
##### Add3(in Thousands) 309
##### Add4(in Thousands) 298
##### Add5(in Thousands) 320

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
x, y = make_regression(n_samples=5, n_features=1, noise=0.1)
model = LinearRegression()
model.fit(x, y)
x = np.array([290,312,309,298,320]).reshape((-1, 1))
print(x)
y = model.predict(x)
print(y)
```
