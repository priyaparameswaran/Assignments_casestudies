# Case Study on Testing of Hypothesis
### A company started to invest in digital marketing as a new way of their product promotions.For that they collected data and decided to carry out a study on it.
###   ● The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.
###   ● The company needs to check whether there is any dependency between the features “Region” and “Manager”.
### Help the company to carry out their study with the help of data provided.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data_set1=pd.read_csv('Sales_add.csv')
data_set1
```

```python
data_set1.describe()
```

### 1. To check whether there is any increase in sales after stepping into digital marketing

```python
print(np.var(data_set1['Sales_before_digital_add(in $)']), np.var(data_set1['Sales_After_digital_add(in $)']))
```

```python
import scipy.stats as stats
stats.ttest_ind(a=data_set1['Sales_before_digital_add(in $)'], b=data_set1['Sales_After_digital_add(in $)'], equal_var=True)
```

### Insights:
##### Here, the P value (= 2.614368006904645e-16) is much less than 0.05, we can assume that there is difference in sales after stepping to Digital Marketing. Thus we can reject Null Hypothesis

### 2. To check whether there is any dependency between the features “Region” and “Manager”

##### The dependency can be figured out by doing a hypothesis testing.If the testing results in a Null hypothesis,the given variables, Region and Manager will be independent of each other.
##### If the data counts for Alternate hypothesis, the variables will be dependent.

```python
from scipy.stats import chi2_contingency
data_R = data_set1['Region']
data_M = data_set1['Manager']
data_crosstable= pd.crosstab(data_R,data_M)
data_crosstable
```

```python
chiValue,pValue,dot,exp=chi2_contingency(data_crosstable)
chiValue,pValue,dot,exp
```

### Insights:
##### Here, the chi-square value obtained is 3.051 and the P-value is 0.5493991051158094, the P-value is much greater than 0.05, null hypothesis cannot be rejected.This means that no relationship exists on the categorical variables, Region and Manager; they are independent.
