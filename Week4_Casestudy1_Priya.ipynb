### Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(8,10))
cars_data1.head()
female=0
male=0
cars_data1.head()
female=list(cars_data1['Buyer Gender']).count('Female')
male=list(cars_data1['Buyer Gender']).count('Male')
gender = ['Male','Female']
count= [male,female]
plt.ylim(4500, 5500)
plt.bar(gender,count, width=0.5)
plt.title("Male and Female Buyers comparison")
plt.xlabel("Buyer Gender")
plt.ylabel("Number of Male/Female")
plt.grid()
plt.show()
```

##### Insights:
##### The above data shows the comparison of Male and Female Buyers. It is shocking to see that the Number of Female Buyers are more when compared to Male Buyers! 
##### The total number of Female Buyers is 5052 and Male Buyers is 4948.
