## 3. Check whether all the zones generate the same amount of sales.

#### Here, we have one independent variable, Amount of sales, thus we can analyze using One way ANOVA.
#### The below are the desired results:
#### Null Hypothesis:Same amount of sales for all Zones
#### Alternate Hypothesis:Not the same amount of sales for all Zones

```python
import scipy
from scipy.stats import f_oneway
import scipy.stats as stats
fvalue, pvalue = stats.f_oneway(data_set1['Zone - A'],data_set1['Zone - B'],data_set1['Zone - C'],data_set1['Zone - D'])
print('The F value is', fvalue)
print('The p-value is', pvalue)
```

## Insights:
#### The p value obtained from ANOVA Test is 0.0011827601694503335 which is pretty less than 0.05, therefore, we can conclude that the Zones generate different amount of sales. Thus Alternate Hypothesis is applicable here.
